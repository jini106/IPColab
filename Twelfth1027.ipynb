```python
img=cv2.imread('/content/drive/MyDrive/IP/Images/Images/lena256.jpg')

print(img.shape)
gImg1=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
#plt.imshow(gImg1,cmap='gray')
cv2_imshow(gImg1)
```

```python
def saturation(value):
    if(value>255):
        value = 255
    elif(value<0):
        value = 0
    return value
```

```python
kernel3 = np.ones((3, 3), np.float32) / 9
kernel5 = np.ones((5, 5), np.float32) / 25
kernel7 = np.ones((7, 7), np.float32) / 49
```

```python
output = cv2.filter2D(gImg1, -1, kernel3)
cv2_imshow(output)
```

```python
output = cv2.filter2D(output, -1, kernel3)
cv2_imshow(output)
```

```python
output = cv2.filter2D(gImg1, -1, kernel5)
cv2_imshow(output)
```

```python
output = cv2.filter2D(output, -1, kernel7)
cv2_imshow(output)
```

```python
gOut = np.zeros((gImg1.shape[0], gImg1.shape[1]), dtype=np.ubyte)
```

```python
for h in range(1, gImg1.shape[0] - 1):
    for w in range(1, gImg1.shape[1] - 1):
        sum = 0
        for i in range(-1, 2):
            for j in range(-1, 2):
                sum+=gImg1[h + i, w + j] * kernel3[i + 1, j + 1]
            sum=saturation(sum)
        gOut[h, w]=sum

cv2_imshow(gOut)
```

```python
kernel3[0][0] = -1 / 9
kernel3[0][1] = -1 / 9
kernel3[0][2] = -1 / 9
kernel3[1][0] = -1 / 9
kernel3[1][1] = 8 / 9
kernel3[1][2] = -1 / 9
kernel3[2][0] = -1 / 9
kernel3[2][1] = -1 / 9
kernel3[2][2] = -1 / 9
```

```python
output = cv2.filter2D(gImg1, -1, kernel3)
output = output * 5
plt.imshow(output, cmap = 'gray')

#cv2_imshow(output)
```
